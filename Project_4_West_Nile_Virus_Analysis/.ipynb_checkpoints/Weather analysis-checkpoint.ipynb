```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from shapely.geometry import Point, Polygon
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

import plotly.express as px
import plotly.graph_objects as go

```

Analysing the presence of the West Nile Virus by location. 

```python
# Create year and month column
train['year'] = pd.DatetimeIndex(train['Date']).year
train['month'] = pd.DatetimeIndex(train['Date']).month
```

```python
train.head()
```

```python
# Count of positive West Nile Virus per year
train.groupby(['year'])['WnvPresent'].sum()
```

```python
# West Nile Virus count for each year by month
sns.pointplot(data=train, x="month", y="WnvPresent", hue="year");
```

The virus peaks in August which is typically the hottest month during Summer. 

```python
# Streets that are more likely to have infected mosquitoes

year = train['year']
x = train.groupby([year, 'Block', 'Street'])['WnvPresent'].sum()
y = train.groupby([year, 'Block', 'Street'])['Street'].count()

z = pd.concat([x, y], axis=1)
z['% Count'] = np.round(x/y, 4)*100

z.index.rename(['Year','Block','StreetName'],inplace=True)
z.reset_index(inplace=True, )
z.set_index(['Block','StreetName'], inplace=True)
z.rename({'Stree':'Strt_Rec_Cnt'}, inplace=True)
z[z['% Count']>0].sort_values(by=['Year','% Count'], ascending=[True,False])
```

```python
z.index = ['{}_{}'.format(i, j) for i, j in z.index]
```

```python
fig, ax = plt.subplots(2,2, figsize = (20,25))
chicago_map = gpd.read_file('data/map/geo_export_c1136ecf-9473-4a42-8d0e-524480fe2323.shp')

# Converting data into right format by
# Converting pandas df into a geo_df ( original_df, co-ordinate_reference_system, geometry)
crs = {'init': 'epsg:4326'}
geometry = [Point(xy) for xy in zip(train['Longitude'],train['Latitude'])]
geo_df = gpd.GeoDataFrame(train, crs=crs, geometry=geometry)

# Plotting the data from GeoDataFrame to our original map.
# 1. 2007
chicago_map.boundary.plot(ax=ax[0,0], alpha=0.6, color='grey')
geo_df[(geo_df['WnvPresent']==0) & (geo_df['year']==2007)].plot(color='green', ax =ax[0,0], label= 'Wnv -' )
geo_df[geo_df['WnvPresent']==1 & (geo_df['year']==2007)].plot(color='red', ax =ax[0,0], label= 'Wnv +')
ax[0,0].set_title('Year: 2007')
ax[0,0].legend(loc='upper right')
plt.legend();

# # 2. 2009
chicago_map.boundary.plot(ax=ax[0,1], alpha=0.6, color='grey')
geo_df[(geo_df['WnvPresent']==0) & (geo_df['year']==2009)].plot(color='green', ax =ax[0,1], label= 'Wnv -' )
geo_df[geo_df['WnvPresent']==1 & (geo_df['year']==2009)].plot(color='red', ax =ax[0,1], label= 'Wnv +')
ax[0,1].set_title('Year: 2009')
ax[0,1].legend(loc='upper right')
plt.legend();

# 3. 2011
chicago_map.boundary.plot(ax=ax[1,0], alpha=0.6, color='grey')
geo_df[(geo_df['WnvPresent']==0) & (geo_df['year']==2011)].plot(color='green', ax =ax[1,0], label='Wnv -' )
geo_df[geo_df['WnvPresent']==1 & (geo_df['year']==2011)].plot(color='red', ax =ax[1,0], label= 'Wnv +')
ax[1,0].set_title('Year: 2011')
ax[1,0].legend(loc='upper right')
plt.legend();

# 4. 2013
chicago_map.boundary.plot(ax=ax[1,1], alpha=0.6, color='grey')
geo_df[(geo_df['WnvPresent']==0) & (geo_df['year']==2007)].plot(color='green', ax =ax[1,1], label= 'Wnv -' )
geo_df[geo_df['WnvPresent']==1 & (geo_df['year']==2007)].plot(color='red', ax =ax[1,1], label= 'Wnv +')
ax[1,1].set_title('Year: 2013')
ax[1,1].legend(loc='upper right')
plt.legend();
```

```python
fig, ax = plt.subplots(1,1, figsize = (20,5))
sns.pointplot( x = z[(z['% Count']>5) & (z['Year']==2007)].index, y = z[(z['% Count']>5) & (z['Year']==2007)]['% Count'], color='g', ax=ax, label= '2007');
sns.pointplot( x = z[(z['% Count']>5) & (z['Year']==2009)].index, y = z[(z['% Count']>5) & (z['Year']==2009)]['% Count'], color='b', ax=ax, label= '2009');
sns.pointplot( x = z[(z['% Count']>5) & (z['Year']==2011)].index, y = z[(z['% Count']>5) & (z['Year']==2011)]['% Count'], color='r', ax=ax, label= '2011');
sns.pointplot( x = z[(z['% Count']>5) & (z['Year']==2013)].index, y = z[(z['% Count']>5) & (z['Year']==2013)]['% Count'], color='y', ax=ax, label= '2013');
ax.tick_params('x',labelrotation=90, labelsize='small', );
ax.set_title('Virus spread distribution in blocks and their streets in 2007, 2009, 2011 & 2013', fontdict = {'fontsize':20});
ax.legend(); 
```

The virus presence is mostly dominated in the year 2007 and 2013. One other interesting thing is for all those blocks/street where the virus count was low in 2007 have gone high in the year 2013.

Analysing impact of temperature on mosquitoes and virus. Features containing temperatues are Tmin, Tmax, and Tavg.

```python
# Look at temperature against dates
temp_date = train.groupby(['Date'], as_index=False)[['Tavg']].mean()
```

```python
# Number of mosquitoes against date
mos_date = train.groupby(['Date'], as_index = False)[['NumMosquitos']].sum()
```

```python
# West nile virus against date
wnv_date = train.groupby(['Date'], as_index = False)[['WnvPresent']].sum()
```

```python
# Merge the 3 df
wnv_mos = pd.merge(mos_date, wnv_date, on = 'Date')
temp_mos_df = pd.merge(wnv_mos, temp_date, on = 'Date')
```

```python
temp_mos_df
```

```python
# Scatter plot showing temperature against the number of mosquitoes and with west nile virus
fig_mos = plt.scatter(x= temp_mos_df['Tavg'], y= temp_mos_df['NumMosquitos'], s= temp_mos_df['WnvPresent'], alpha= 0.7, edgecolors='white', linewidth=1 );
```

Scatterplot shows that generally with higher temperatures indicate higher number of mosquitoes and therefore an increased presence of the virus. 

```python
# Effect of temperature.
# Let us see the distribution of temperatures for min, max and avg wrt. the mosquito virus infection.

fig, ax = plt.subplots(1,4, figsize=(20,5))
sns.distplot(train['Tmin'], ax=ax[0])
ax[0].xaxis.set(ticks=np.arange(40,80,5))
ax[0].set_title('Tmin')

sns.distplot(train['Tmax'], ax=ax[1])
ax[1].xaxis.set(ticks=np.arange(50,100,5))
ax[1].set_title('Tmax')

sns.distplot(train['Tavg'].apply(int), ax=ax[2])
ax[2].xaxis.set(ticks=np.arange(40,90,5))
ax[2].set_title('Tavg')

sns.distplot(train['Depart'].apply(int), ax=ax[3])
ax[3].set_title('Depart');
```

The data shows that there is a positive correlation between higher temperatures and increased presence of the virus. There are also multiple peaks which could potentially indicate there may be a presence of some other phenomenon. 

```python
fig, ax = plt.subplots(1,1, figsize=(20,5))
sns.distplot(train.loc[train['year']==2007,'Tavg'].apply(int), ax=ax,color='g',hist=False, label= '2007')
sns.distplot(train.loc[train['year']==2009,'Tavg'].apply(int), ax=ax,color='b',hist=False, label= '2009')
sns.distplot(train.loc[train['year']==2011,'Tavg'].apply(int), ax=ax,color='r',hist=False, label= '2011')
sns.distplot(train.loc[train['year']==2013,'Tavg'].apply(int), ax=ax,color='y',hist=False, label= '2013')
ax.xaxis.set(ticks=np.arange(40,90,5))
ax.set_title('Tavg')
plt.legend();
```

From the above we can see that 4 peaks we see in the avg temp distribution was due to high temperatures in different years.

 * 2007 - The mean temp was between 75-80.
 * 2009 - The mean temp was between 65-70 and the peak hike was attained during 70-75 range.
 * 2011 - The mean temp was 75 but we could notice a small bump in the early 50-55 range as well.
 * 2013 - The max hike was obtained during 75-80 range.

This explains why we have 4 peaks in our Tavg overall plot above

```python
# Moisture content in Atmosphere using DewPoint & WetBulb test
# Let us plot the data distribution and how are variable is distributed in the space.
fig, ax = plt.subplots(1,2, figsize=(20,5))
sns.distplot(train['DewPoint'], ax=ax[0])
ax[0].set_title('DewPoint');

sns.distplot(train['WetBulb'], ax=ax[1])
ax[1].set_title('WetBulb');
```

Dew Point - The dew point is the temperature to which air must be cooled to become saturated with water vapor. So if the air is cooling at a higher temp that means it has already enough of moisture present and it just needs to be cooled by a very minimal amount before it starts condensing into water vapour.

Wet Bulb - Similar is the case with Wet Bulb. If the Wet Bulb temp is significantly lower then it means the air earlier was pretty dry and vice-versa.

Based on the above observations we could say that lower the temperature variance from the Tavg more is the humidity already present in the atmosphere.

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(20,5));
sns.violinplot(x = train['year'], y = train['Tavg']-train['DewPoint'], hue = train['WnvPresent'], ax=ax1);
ax1.set_title('Humidity Estimate from DewPoint', fontdict= {'fontsize':15});
sns.violinplot(x = train['year'], y = train['Tavg']- train['WetBulb'], hue = train['WnvPresent']);
ax2.set_title('Humidity Estimate from Wet Bulb Temperature', fontdict= {'fontsize':15});
ax2.set_ylim(0,20);
```

From the above plot we can see that for Virus to be present the departure from the average temperature is not much (except for 2013 where it shows equal for both) which indicates the air was already containing some moisture in it.

The plot above confirms our assumption even further. For almost all the years the temp distribution is on the higher side where Wnv is present.
